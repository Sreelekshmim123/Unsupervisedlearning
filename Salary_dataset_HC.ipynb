```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
ss.fit_transform(data)
```

```python
import scipy.cluster.hierarchy as sch
x=sch.dendrogram(sch.linkage(x1,method='ward'))
plt.title("Dendrogram")
plt.xlabel("years Experience")
plt.ylabel("salary")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
hc=AgglomerativeClustering(n_clusters=2,affinity='euclidean',linkage='ward')
yhc=hc.fit_predict(x1)
```

```python
yhc
```

```python
plt.scatter(x1[yhc==0,0],x1[yhc==0,1],c="red",label="Cluster 1")
plt.scatter(x1[yhc==1,0],x1[yhc==1,1],c="yellow",label="Cluster 2")
plt.title("Salary perdiction")
plt.xlabel("years of expereience")
plt.ylabel("Salary")
plt.show()
```

```python
from sklearn.metrics import silhouette_score
sc=silhouette_score(x1,yhc)
sc
```
