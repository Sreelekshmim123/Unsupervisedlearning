```python
df = pd.read_csv("/content/Mall_Customers.csv")
df.head()
```

```python
x =df.iloc[:,[3,4]].values
x
```

```python
import scipy.cluster.hierarchy as sch
dentrogram =sch.dendrogram(sch.linkage(x,method='ward'))
plt.title('Dendrogram')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
hc=AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')
yhc=hc.fit_predict(x)
```

```python
yhc
```

```python
plt.scatter(x[yhc==0,0],x[yhc==0,1],c="red",label="Cluster 1")
plt.scatter(x[yhc==1,0],x[yhc==1,1],c="yellow",label="Cluster 2")
plt.scatter(x[yhc==2,0],x[yhc==2,1],c="blue",label="Cluster 3")
plt.scatter(x[yhc==3,0],x[yhc==3,1],c="orange",label="Cluster 4")
plt.scatter(x[yhc==4,0],x[yhc==4,1],c="grey",label="Cluster 5")
plt.title("Clusters of Customers")
plt.xlabel("Annual income")
plt.ylabel("Spending score")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import silhouette_score
sc=silhouette_score(x,yhc)
sc
```
