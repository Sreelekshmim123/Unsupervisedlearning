**KMEANS ALGORITHM**

K-Means Clustering is an unsupervised learning algorithm that is used to solve the
clustering problems in machine learning or data science.K-Means Clustering is an
Unsupervised Learning algorithm, which groups the unlabeled dataset into different
clusters.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from google.colab import files
uploaded  = files.upload()
```

```python
df  = pd.read_csv("/content/Mall_Customers.csv")
df.head()
```

```python
x =df[["Annual Income (k$)","Spending Score (1-100)"]]
x.head()
```

```python
from sklearn.cluster import KMeans
```

```python
from os import initgroups
wcss =[]
for i in range(1,11):
  kmeans  =KMeans(n_clusters=i,init='k-means++',max_iter=300,n_init=10,random_state=1)
  kmeans.fit(x)
  wcss.append(kmeans.inertia_)
```

```python
plt.figure(figsize =(10,5))
plt.plot(range(1,11),wcss,marker='o',color='blue')
plt.show()
```

```python
kmeans  =KMeans(n_clusters=5,init='k-means++',max_iter=300,n_init=10,random_state=1)
ymeans= kmeans.fit_predict(x)
```

```python
ymeans
```

```python
df['cluster']=ymeans
df.head()
```

**Ploting** the cluster

```python
plt.scatter(x['Annual Income (k$)'],x['Spending Score (1-100)'],c=ymeans)
plt.show()
```

```python
colors =['r','g','b','y','m']
for i in range(5):
  clusterdata =x[ymeans==i]
  plt.scatter(clusterdata['Annual Income (k$)'],clusterdata['Spending Score (1-100)'],c=colors[i],label=f'Cluster{i}' )
plt.legend()
plt.show()
```

```python
def predict_cluster(annual_income,spending_score):
  newdata=[[annual_income,spending_score]]
  clusters=kmeans.predict(newdata)
  return clusters
```

```python
new_customer_income=80
new_customer_score=20
predictcluster=predict_cluster(new_customer_income,new_customer_score)
print("The customers belongs to the cluster :",predictcluster)
```
